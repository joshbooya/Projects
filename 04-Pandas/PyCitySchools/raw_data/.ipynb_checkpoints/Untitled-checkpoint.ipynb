```python
import pandas as pd
```

```python
data1 = "schools_complete.csv"
data2 = "students_complete.csv"

df1 = pd.read_csv(data1)
df2 = pd.read_csv(data2)
df1 = df1.rename(columns={"name":"school"})
```

```python
#**District Summary**
#
#* Create a high level snapshot (in table form) of the district's key metrics, including:
#  * Total Schools
#  * Total Students
#  * Total Budget
#  * Average Math Score
#  * Average Reading Score
#  * % Passing Math
#  * % Passing Reading
#  * Overall Passing Rate (Average of the above two)

```

```python
df1
```

```python
dfm = pd.merge(df1,df2, on="school")
```

```python
ttl_schools = dfm.school.nunique()
ttl_students = dfm["Student ID"].nunique()
ttl_budget = dfm.budget.sum()
avg_math = dfm.math_score.mean()
avg_read = dfm.reading_score.mean()
pm = dfm[dfm.math_score >= 65]["size"].count()/ttl_students
pass_math = "{:,.2f}%".format(pm*100)
pr = dfm[dfm.reading_score >= 65]["size"].count()/ttl_students
pass_read = "{:,.2f}%".format(pr*100)
pass_rate = "{:,.2f}%".format(((pr+pm)/2)*100)

pd.DataFrame({"Total Schools": [ttl_schools],"Total Students": [ttl_students],"Total Budget": [ttl_budget],"Average Math Score": [avg_math],"Average Reading Score": [avg_read],"% Passing Math": [pass_math],"% Passing Reading": [pass_read],"Overall Passing Rate (Average of the above two)": [pass_rate]})
```

```python
sum_df = dfm[["school","type","size","budget","reading_score","math_score"]]

df_type_students_budget = df1[["school","type","size","budget"]]
df_type_students_budget["Per Student Budget"] = df_type_students_budget["budget"]/df_type_students_budget["size"]

sum_avg_read = sum_df.groupby("school")["reading_score"].mean()
sum_avg_math = sum_df.groupby("school")["math_score"].mean()
sum_pass_math = sum_df[sum_df.math_score>=65].groupby("school")["math_score"].count()/sum_df.groupby("school")["size"].count()
sum_pass_read = sum_df[sum_df.reading_score>=65].groupby("school")["reading_score"].count()/sum_df.groupby("school")["size"].count()
ttl_budget = dfm[[]]
```
