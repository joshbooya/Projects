```python
import os
import csv

csvpath1 = 'budget_data_1.csv'
csvpath2 = 'budget_data_2.csv'

with open(csvpath1, 'r') as csvfile1:
    # Split the data on commas
    csvreader1 = csv.reader(csvfile1, delimiter=',')
    next(csvreader1)
    # counts up the months
    month1 = 0
    totalrev1 = 0
    monthly_change1 = 0
    monthly_sum1 = 0
    avgrevchange1 = 0
    rev_change1 = []
    rev_array1 = []
    great_dec1 = 0
    great_mo1 = 'string'
    great_inc1 = 0
    lowest_mo1 = 'string'

    for row in csvreader1:
        revenue = float(row[1])
        rev_array1.append(revenue)
        totalrev1 += revenue
        month1 += 1
        if revenue < great_dec1:
            great_dec1 = revenue
            great_mo1 = row[0]
        elif revenue > great_inc1:
            great_inc1 = revenue
            lowest_mo1 = row[0]

    for i in range(len(rev_array1)-1):
        monthly_change1 = rev_array1[i+1]-rev_array1[i]
        monthly_sum1 += monthly_change1
        rev_change1.append(monthly_change1)
    avgrevchange1 = monthly_sum1/len(rev_change1)


with open(csvpath2, 'r') as csvfile2:
    # Split the data on commas
    csvreader2 = csv.reader(csvfile2, delimiter=',')
    next(csvreader2)
    # counts up the months
    month2 = 0
    totalrev2 = 0
    monthly_change2 = 0
    monthly_sum2 = 0
    avgrevchange2 = 0
    rev_change2 = []
    rev_array2 = []
    great_dec2 = 0
    great_mo2 = 'string'
    great_inc2 = 0
    lowest_mo2 = 'string'

    for row2 in csvreader2:
        revenue = float(row2[1])
        rev_array2.append(revenue)
        totalrev2 += revenue
        month2 += 1
        if revenue < great_dec2:
            great_dec2 = revenue
            great_mo2 = row2[0]
        elif revenue > great_inc2:
            great_inc2 = revenue
            lowest_mo2 = row2[0]

    for i in range(len(rev_array2)-1):
        monthly_change2 = rev_array2[i+1]-rev_array2[i]
        monthly_sum2 += monthly_change2
        rev_change2.append(monthly_change2)
    avgrevchange2 = monthly_sum2/len(rev_change2)

    print ("For budget_data_1.csv " + '\n' + 
        "Total Months: " + str(month1) + '\n' + 
        "Total Revenue: " + str(totalrev1) + '\n' + 
        "Average Revenue Change: " + str(avgrevchange1) + '\n' + 
        "Greatest Increase in Revenue: " + great_mo1 + " (" + str(great_inc1) + ") " + '\n' + 
        "Greatest Decrease in Revenue: " + lowest_mo1 + " (" + str(great_dec1) + ") " + '\n' + '\n' +
        "For budget_data_2.csv " + '\n' + 
        "Total Months: " + str(month2) + '\n' + 
        "Total Revenue: " + str(totalrev2) + '\n' + 
        "Average Revenue Change: " + str(avgrevchange2) + '\n' + 
        "Greatest Increase in Revenue: " + great_mo2 + " (" + str(great_inc2) + ") " + '\n' + 
        "Greatest Decrease in Revenue: " + lowest_mo2 + " (" + str(great_dec2) + ") ") 
```
