```python
import pandas as pd

#FINDINGS
#1. Both data points, reveal that Males generally play the game more than Females.
#2. The largest age group playing this game are 20-24 year olds. They are also the largest purchasers in the game.
#3. The weapons that made the largest profits was the "Retribution Axe" and the "Mourning Blade."

purchase1 = "purchase_data.json"
purchase2 = "purchase_data2.json"
df1 = pd.read_json(purchase1)
df2 = pd.read_json(purchase2)
```

```python
#**Player Count**
#* Total Number of Players
players1 = df1.SN.nunique()
players2 = df2.SN.nunique()
Player_Counts = pd.DataFrame({"Player Count":[players1, players2]}, index = ["File 1","File 2"])
#ALTERNATIVE: print(f" Total Number of Players: \n File 1 - {players1} \n File 2 - {players2}")
Player_Counts
```

```python
#**Purchasing Analysis (Total)**

#* Number of Unique Items
no_unique_items1 = df1["Item Name"].nunique()
no_unique_items2 = df2["Item Name"].nunique()

#* Average Purchase Price
avg_purch_price1 = df1.Price.mean()
avg_purch_price2 = df2.Price.mean()

#* Total Number of Purchases
ttl_no_purch1 = df1.SN.count()
ttl_no_purch2 = df2.SN.count()

#* Total Revenue
ttl_rev1 = df1.Price.sum()
ttl_rev2 = df2.Price.sum()

#RESPONSE DATAFRAME
Purchasing_Analyis = pd.DataFrame({"Total Number of Purchases":[ttl_no_purch1, ttl_no_purch2] ,"Average Purchase Price":[avg_purch_price1, avg_purch_price2],"Number of Unique Items": [no_unique_items1, no_unique_items2]},index = ["File 1","File 2"])
Purchasing_Analyis = Purchasing_Analyis.style.format({"Average Purchase Price": lambda x: "${:,.2f}".format(x)})
Purchasing_Analyis
```

```python
#**Age Demographics**
#
#* The below each broken into bins of 4 years (i.e. &lt;10, 10-14, 15-19, etc.) 
#  * Purchase Count
#  * Average Purchase Price
#  * Total Purchase Value
#  * Normalized Totals

#AGE DEMOS - DATA SET 1
ag1 = pd.DataFrame(df1["Age Group"].value_counts()[binlabel])
ag1 = ag1.rename(columns={"Age Group":"Purchase Count I"})
ag1["Average Purchase Price I"] = df1.groupby(["Age Group"])["Price"].mean()
ag1["Total Purchase Value I"] = df1.groupby(["Age Group"])["Price"].sum()
#ag1.style.format({'Average Purchase Price I':"${:,.2f}"})
age_demo1 = ag1.style.format({"Average Purchase Price I": lambda x: "${:,.2f}".format(x), "Total Purchase Value I": lambda x: "${:,.2f}".format(x)})
age_demo1
```

```python
#AGE DEMOS - DATA SET 2
ag2 = pd.DataFrame(df2["Age Group"].value_counts()[binlabel])
ag2 = ag2.rename(columns={"Age Group":"Purchase Count II"})
ag2["Average Purchase Price II"] = df2.groupby(["Age Group"])["Price"].mean()
ag2["Total Purchase Value II"] = df2.groupby(["Age Group"])["Price"].sum()
#ag2.style.format({'Average Purchase Price':"${:,.2f}"})
age_demo2 = ag2.style.format({"Average Purchase Price II": lambda x: "${:,.2f}".format(x), "Total Purchase Value II": lambda x: "${:,.2f}".format(x)})
age_demo2
```

```python
#**Top Spenders**
#
#* Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
#  * SN
#  * Purchase Count
#  * Average Purchase Price
#  * Total Purchase Value

top5_1SN = df1.SN.value_counts()
SN1 = pd.DataFrame(top5_1SN.head(5))
SN1 = SN1.rename(columns={"SN":"Count"})
SN1_list = list(SN1.index)

    #Output: ['Undirrala66', 'Saedue76', 'Qarwen67', 'Hailaphos89', 'Sondastan54']

#New Data Frame for analysis
SN1_aa = df1[df1["SN"] == SN1_list[0]]
SN1_ab = df1[df1["SN"] == SN1_list[1]]
SN1_ac = df1[df1["SN"] == SN1_list[2]]
SN1_ad = df1[df1["SN"] == SN1_list[3]]
SN1_ae = df1[df1["SN"] == SN1_list[4]]
SN_frames1 = [SN1_aa, SN1_ab, SN1_ac, SN1_ad, SN1_ae]
result1 = pd.concat(SN_frames1)
#result1

#Relabeling columns
Purchase_Count1 = pd.DataFrame(result1.groupby(["SN"])["Price"].count())
Average_Purchase_Price1 = pd.DataFrame(result1.groupby(["SN"])["Price"].mean())
Average_Purchase_Price1 = Average_Purchase_Price1.rename(columns={"Price":"Average Purchase Price"})
Total_Purchase_Value1 = pd.DataFrame(result1.groupby(["SN"])["Price"].sum())
Total_Purchase_Value1 = Total_Purchase_Value1.rename(columns={"Price":"Total Purchase Value"})

#TOP SPENDERS - DATA SET 1
Top_Spenders1 = pd.merge(pd.merge(Purchase_Count1,Average_Purchase_Price1,left_index=True,right_index=True),Total_Purchase_Value1,left_index=True,right_index=True).rename(columns={"Price":"Purchase Count"})
Top_Spenders1 = Top_Spenders1.rename(columns={"Purchase Count": "Purchase Count I","Average Purchase Price": "Average Purchase Price I","Total Purchase Value": "Total Purchase Value I"})
Top_Spenders1 = Top_Spenders1.style.format({"Average Purchase Price I": lambda x: "${:,.2f}".format(x), "Total Purchase Value I": lambda x: "${:,.2f}".format(x)})
Top_Spenders1
```

```python
#**Most Popular Items**
#
#* Identify the 5 most popular items by purchase count, then list (in a table):
#  * Item ID
#  * Item Name
#  * Purchase Count
#  * Item Price
#  * Total Purchase Value
top5_1Item = df1["Item ID"].value_counts()
Item1 = pd.DataFrame(top5_1Item.head(5))
Item1 = Item1.rename(columns={"Item ID":"Purchase Counts"})
Item1_list = list(Item1.index)
#Item1_list

#New Data Frame for analysis
Item1_aa = df1[df1["Item ID"] == Item1_list[0]]
Item1_ab = df1[df1["Item ID"] == Item1_list[1]]
Item1_ac = df1[df1["Item ID"] == Item1_list[2]]
Item1_ad = df1[df1["Item ID"] == Item1_list[3]]
Item1_ae = df1[df1["Item ID"] == Item1_list[4]]
Item_frames1 = [Item1_aa, Item1_ab, Item1_ac, Item1_ad, Item1_ae]
Item_result1 = pd.concat(Item_frames1)
Item_result1 = Item_result1.reset_index(drop=True)
#Item_result1 = Item_result1.groupby(["Item ID","Item Name"])["Price"].unique()

#Relabeling columns
Item_Purchase_Count1 = pd.DataFrame(Item_result1.groupby(["Item ID","Item Name"])["Price"].count())
Item_Purchase_Count1 = Item_Purchase_Count1.rename(columns={"Price":"Purchase Count"})
#Item_Price1 = Item_result1[["Price"]].groupby(["Item ID","Item Name"])
#Item_Price1 = Item_Price1.rename(columns={"Price":"Item Price"})
Item_Total_Purchase_Value1 = pd.DataFrame(Item_result1.groupby(["Item ID","Item Name"])["Price"].sum())
Item_Total_Purchase_Value1 = Item_Total_Purchase_Value1.rename(columns={"Price":"Total Purchase Value"})
Item_Price1 = Item_result1[["Item ID","Item Name","Price"]].drop_duplicates().reset_index(drop=True)
Item_Price1 = Item_Price1.set_index(["Item ID","Item Name"]).rename(columns={"Price":"Item Price"})

#MOST POPULAR ITEMS - DATA SET 1
Pop_Items1 = pd.merge(pd.merge(Item_Purchase_Count1,Item_Price1,left_index=True,right_index=True),Item_Total_Purchase_Value1,left_index=True,right_index=True)
Pop_Items1 = Pop_Items1.rename(columns={"Purchase Count": "Purchase Count I","Item Price": "Item Price I","Total Purchase Value": "Total Purchase Value I"})
Pop_Items1 = Pop_Items1.style.format({"Item Price I": lambda x: "${:.2f}".format(x), "Total Purchase Value I": lambda x: "${:.2f}".format(x)})
Pop_Items1
```
