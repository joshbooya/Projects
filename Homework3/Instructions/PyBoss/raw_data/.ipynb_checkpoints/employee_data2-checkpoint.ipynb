```python
import os
import csv

csvpath1 = 'employee_data2.csv'
emp_ID = []
Name = []
First_Name = [] 
Last_Name = []
DOB = []
full_SSN = []
SSN = []
full_State = []
State = []
#a = []
#b = []
YMD = [] #Year Month Day
d1 = []
e1 = []
f1 = []
st = []
us_state_abbrev = {
    'Alabama': 'AL',
    'Alaska': 'AK',
    'Arizona': 'AZ',
    'Arkansas': 'AR',
    'California': 'CA',
    'Colorado': 'CO',
    'Connecticut': 'CT',
    'Delaware': 'DE',
    'Florida': 'FL',
    'Georgia': 'GA',
    'Hawaii': 'HI',
    'Idaho': 'ID',
    'Illinois': 'IL',
    'Indiana': 'IN',
    'Iowa': 'IA',
    'Kansas': 'KS',
    'Kentucky': 'KY',
    'Louisiana': 'LA',
    'Maine': 'ME',
    'Maryland': 'MD',
    'Massachusetts': 'MA',
    'Michigan': 'MI',
    'Minnesota': 'MN',
    'Mississippi': 'MS',
    'Missouri': 'MO',
    'Montana': 'MT',
    'Nebraska': 'NE',
    'Nevada': 'NV',
    'New Hampshire': 'NH',
    'New Jersey': 'NJ',
    'New Mexico': 'NM',
    'New York': 'NY',
    'North Carolina': 'NC',
    'North Dakota': 'ND',
    'Ohio': 'OH',
    'Oklahoma': 'OK',
    'Oregon': 'OR',
    'Pennsylvania': 'PA',
    'Rhode Island': 'RI',
    'South Carolina': 'SC',
    'South Dakota': 'SD',
    'Tennessee': 'TN',
    'Texas': 'TX',
    'Utah': 'UT',
    'Vermont': 'VT',
    'West Virginia': 'WV',
    'Virginia': 'VA',
    'Washington': 'WA',
    'Wisconsin': 'WI',
    'Wyoming': 'WY',
}

# function to replace text using dictionary list
def multipleReplace(text, wordDict):
    for key in wordDict:
        text = text.replace(key, wordDict[key])
    return text

with open(csvpath1, 'r') as csvfile1:
    csv1 = csv.reader(csvfile1, delimiter=',')
    
    for row in csv1:
        emp_ID.append(row[0])
        Name.append(row[1])
        Name[0] = "FName LName" # rewrites 1st cell of Name as two item list
        YMD.append(row[2])
        YMD[0]="Year-Month-Day"
        full_SSN.append(row[3])
        full_SSN[0]="SSN1-SSN2-SSN3"
        full_State.append(row[4])

        # print(full_SSN)
    
    # handling name as string and splitting the string re-appended to a list 
    
    for r1 in Name:
        First, Last = r1.split(' ')
        First_Name.append(First)
        Last_Name.append(Last)
        First_Name[0] = "First Name"
        Last_Name[0] = "Last Name"
    for r2 in YMD:
        a, b, c = r2.split('-')
        mdy = b +"/" + c + "/" + a
        DOB.append(mdy)
        DOB[0] = "DOB"
        # print(type(DOBf))
    for r3 in full_SSN:
        d, e, f = r3.split('-')
        ssn1 = "***-**-" + f
        SSN.append(ssn1)
        SSN[0] = "SSN"
    # running function to update list using 
    for r4 in full_State:
        State.append(multipleReplace(r4,us_state_abbrev))
    print (type(State))
    
        
#    print(us_state_abbrev["KS"]) 
    roster = zip(emp_ID, First_Name, Last_Name, DOB, SSN, State)
    for employee in roster:
        print(employee)
```
